```
上衣 = int(input('請輸入上衣數量?'))
褲子 = int(input('請輸入褲子數量?'))
背心 = int(input('請輸入背心數量?'))
總金額 = 上衣*300 + 褲子*350 + 背心*400
print('訂購服裝的總金額為', 總金額)
```

```
半徑 = float(input('請輸入半徑?'))
PI = 3.14159
圓周長 = 2 * PI * 半徑
圓面積 = 半徑 * 半徑 * PI
print('圓周長為', 圓周長, '圓面積為', 圓面積)
```
